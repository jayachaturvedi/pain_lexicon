```python
# remove whitespaces
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
from gensim.parsing.preprocessing import strip_multiple_whitespaces
```

```python
for text in df:
    gensim.parsing.preprocessing.strip_multiple_whitespaces(text)
```
